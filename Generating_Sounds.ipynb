```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
import IPython
```

```python
def create_waveform(f0, V, h, phi, filename):
    duration = 5 #duration of generated audio file in s
    T = 1/f0
    fs = 44100 #sampling frequency
    t = np.arange(0, T, 1/fs) #Generate waveform for one cycle
    
    y = t
    for i in range(0, len(V)):
        y = y + V[i]*(np.sin(2*np.pi*h[i]*f0*t + phi[i]))
    
    
    plt.plot(t, y)
    plt.show()
    
    y = np.tile(y, int(duration/T)) #Repeat cycle according to duration
    scaled = np.int16(y/np.max(np.abs(y)) * 32767)
    
    write(filename, 44100, scaled) #Generate audio file
```

```python
#tuning fork
f0 = 440
V = [1]
h = [1]
phi = [0]
filename = "tuningfork.wav"

y = create_waveform(f0, V, h, phi, filename)

IPython.display.Audio(filename)
```
